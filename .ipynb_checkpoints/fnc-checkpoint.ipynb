```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b
def cal():
    a=int(input("Enter a = "))
    b=int(input("Enter b= "))
    o=input("Enter operator= ")
    if o=='+':
        print(add(a,b))
    elif o=='-':
        print(sub(a,b))
    elif o=='*':
        print(mul(a,b))
    elif o=='/':
        print(div(a,b))
    else:
        print("Invalid Operator")
    x=input("Enter y for more calculations: ")
    if x=='y':
        cal()
cal()  
```

```python
bill=""
n=int(input("Enter n= "))
def billing():
    global bill
    product=input("Enter product name= ")
    price=int(input("Enter price= "))
    quantity=int(input("Enter quantity= "))
    total= price*quantity
    info=f"{product} {price} {quantity} {total}\n"
    bill=bill+info
for i in range(n):
    billing()
    
print(bill)
```

```python
bill=""
grand_total=0
n=int(input("Enter n= "))
def billing(product,price,quantity,total):
    global bill,grand_total
    info=f"{product} {price} {quantity} {total}\n"
    bill=bill+info
    grand_total=grand_total+total
    
for i in range(n):
    product=input("Enter product name= ")
    price=int(input("Enter price= "))
    quantity=int(input("Enter quantity= "))
    total= price*quantity
    billing(product,price,quantity,total)
    
print("Bill:\n",bill)
print("Grand Total =",grand_total)
```
